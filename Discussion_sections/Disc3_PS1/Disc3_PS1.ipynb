## 1.2 Create relative frequency table
- The same data as above but turned into a percentage

### 1.2.1 Create table: DPLYR
- DPLYR-Count-mutate

```r
mtcars %>% 
count(vs) %>% 
mutate(
    #Percentage is a name you can change! 
    #Similar to how we name our dataframes and other objects
    # n= number of observations by category
    # sum(n)= total number of observations
    Percentage = n/sum(n) 
      )
```

### 1.2.2 Create table: table function

```r
table(mtcars$vs)/length(mtcars$vs) #Divides entire table by the number of observations
```

# 4. Groupby

## 4.1 Summary stats: dplyr--groupby

```r
mtcars %>% 
group_by(am) %>% 
summarise(
    avg_wt = mean(wt),
    sd_wt = sd(wt))
```

## 4.2 Summary stats: mean, sd function

```r
#mean
manual_mean =  mean(mtcars$wt[mtcars$am==1])
#sd
manual_sd = sd(mtcars$wt[mtcars$am==1])
#Put both results in the same table
summary_table = cbind(manual_mean, manual_sd)
summary_table
```
