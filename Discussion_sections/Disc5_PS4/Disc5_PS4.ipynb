# 2. Confidence interval

$$\bar{x} \pm t\frac{s}{\sqrt{n}}$$

## 2.1 Calculation by hand

### 2.1.1 Find t-score

```r
t = qt(0.95, df=length(mtcars$wt-1)) #t score
t
```

### 2.1.2 Divide standard deviation by sqrt(n)

```r
rhs = sd(mtcars$wt)/sqrt(length(mtcars$wt))
```

### 2.1.3 Combine 2.1.1. & 2.1.2

```r
confidence_int = t*rhs
confidence_int 
```

### 2.1.4 Confidence intervals

```r
x_mean = mean(mtcars$wt)
x_mean
```

```r
x_mean-confidence_int
```

```r
x_mean+x_mean
```

## 2.2 Calculation by user defined function

```r
confidence_interval = function(column, confidence_int){   
    t = qt(confidence_int, length(column)-1) #t score    rhs = sd(column)/sqrt(length(column))
    solu = round(t*rhs,4)
    x_mean = round(mean(column),6) #rounds to two digits
    conf_lower = round(x_mean-solu,6)
    conf_upper = round(x_mean+solu,6)
    
    print(paste('Mean:', x_mean,'+-', solu, sep=''))
    print(paste('Upper bound: ', conf_upper, ' Lower bound: ', conf_lower, sep=''))
}
```

```r
confidence_interval(mtcars$wt, .95)
```

# 4.  T-test: 2-sided
- `PS3 corresponding problem: 1.d`


## 4.1 Compare mean weight of cars with automatic and manual

### 4.1.1 filter weight values for automatic cars

```r
wt_automatic = mtcars %>% 
filter(am==1) %>% 
select(wt) 
```

### 4.1.2 filter weight values for automatic cars

```r
wt_manual = mtcars %>% 
filter(am==0) %>% 
select(wt) 
```

### 4.1.3 Use "t.test()" function

```r
test_solution = t.test(wt_automatic, wt_manual)
test_solution
```

### 4.1.4 Extract p-value from solution 

```r
pvalue = test_solution$p.value 
pvalue 
```

```r
if (pvalue <.05){
    print('Reject the null')
    print('Evidence that the means are statisticaly different')   
}
if(pvalue >.05){
    print('Fail to reject null')
    print('No evidence that means are statisticaly different')   
}
```
