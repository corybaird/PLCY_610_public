# 1. Scatterplot

- `PS2 corresponding problem: 2c`

## 1.1 ggplot library

```r
mtcars %>% 
ggplot(
    aes(x =  qsec,y = drat)
)+
geom_point() + #Places points on chart
xlab('1/4 mile time') + 
ylab('Rear axel ratio') 
```

## 1.2 "plot" function

```r
plot(mtcars$qsec, #x
     mtcars$drat, #y
     xlab="1/4 mile time", 
     ylab="Rear axel ratio",
     main="Answer 2.C",
     pch=19 #Sets dot
    )
```

## 3.2 Histograms

### 3.2.1 "Hist: function

```r
par(mfrow=c(1,2))
hist(mtcars$vs, xlab="", main="Engine shape")
hist(mtcars$gear, xlab="", main="Gears")
```

### 3.2.2 ggplot2, extragrid libraries
- Install gridextra

```r
#install.packages('gridExtra')
library(gridExtra)
```

```r
plot1 = mtcars %>% 
ggplot(aes(x=vs)) +  # x = Column_name
geom_histogram(color="black", fill="white", bins=2)
```

```r
plot2 = mtcars %>% 
ggplot(aes(x=gear)) +  # x = Column_name
geom_histogram(color="black", fill="white", bins=3)
```

```r
grid.arrange(plot1, plot2, ncol=2)
```
