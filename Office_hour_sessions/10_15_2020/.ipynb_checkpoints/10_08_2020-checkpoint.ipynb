# 2.H What is the 95% confidence interval of the sample mean for logreceipts for the samples of n = 5 and n=100? 

Display your answer in the following form:

$$\Bigg[\bar{X}_n - z \cdot \frac{\sigma}{\sqrt{n}}, \bar{X}_n + z \cdot \frac{\sigma}{\sqrt{n}}\Bigg]\tag{Formula 1}$$

### Cory's comments:
- To answer this question we use the population data set
    - In this example the population dataframe is labeled is "df"
    
    
- We only need to calculate the standard deviation of log
    - This is done in the code below in 2.H.1

### 2.H.1 Calculate the standard deviation of log receipts from the population data set
- This will replace $\sigma$ in formula 1

```r
sd(df$receipts_log)
```

## 2.H.1 Answer

- Replace $\sigma$ = .62
- Replace $n$ = number of observations

Answer: n=5
$$\Bigg[\bar{X}_n - 1.96 \cdot \frac{.62}{\sqrt{5}}, \bar{X}_n + 1.96 \cdot \frac{.62}{\sqrt{5}}\Bigg]$$

Answer: n=100
$$\Bigg[\bar{X}_n - 1.96 \cdot \frac{.62}{\sqrt{100}}, \bar{X}_n + 1.96 \cdot \frac{.62}{\sqrt{100}}\Bigg]$$

# 1. Norms

## 1.1 pnorm
- Input actual data to find what percentile it is in!

### 1.1.1 Example: mean = 75, sd = 20, OUR TEST SCORE = 95
- Pnorm will tell us what percentile we are in using our score

```r
acutal_score = 95

pnorm(acutal_score, 
      mean = 75, sd = 20) 
```

```r
example_answer = pnorm(acutal_score, mean = 75, sd = 20) 
print(paste("Our score was in the", round(example_answer,2)*100, "percentile"))
```

#### 1.1.1.1 Visualize example

```r
set.seed(10)

hist(rnorm(1000, mean= 75, sd=20),
    main = "p-norm: score of 95 is better than 84% of students",
    xlab='test score') #Generates data for 1000 test scores with mean=75, sd=20

abline(v = 95, col="red")
```

## 1.2 qnorm
- Input percentile to find out the actual data!

- This is just the opposite of pnorm

### 1.2.1 Example: mean = 75, sd = 20, TEST PERCENTILE = 0.841
- Qnorm will tell us our score using our percentile

```r
percentile_score = 0.841344746068543 #THIS IS THE SAME AS OUR ANSWER IN 1.1.1
qnorm(percentile_score, mean=75, sd=20)
```
